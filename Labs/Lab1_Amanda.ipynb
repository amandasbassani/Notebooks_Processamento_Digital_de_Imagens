## 3) Visualize a imagem e o histograma do nível de cinza da imagem Fig3.15(a)3.jpg. Equalize o histograma. 
Veja a nova imagem e o novo histograma. Observe as diferenças entre as imagens e os histogramas. Por que o histograma não ficou perfeitamente
uniformemente distribuído?

```python
im = plt.imread('/content/drive/MyDrive/PDI_Labs/Lab1-Amanda/Lab-1/Fig3.15(a)3.jpg')
plt.figure(figsize = (5,5))
plt.imshow(im,cmap='gray')
plt.show() #exibe a imagem original
```

```python
from skimage import exposure
im2 = skimage.exposure.equalize_hist(np.double(im))
plt.figure(figsize = (5,5))
plt.imshow(np.uint8(255*im2),cmap='gray')
plt.show() #exibe a imagem equalizada
```

```python
plt.hist(im.ravel(), bins = 256, range=[0,255])
plt.show() #exibe o histograma da imagem original

plt.hist(np.uint8(255*im2).ravel(), bins = 256, range=[0,255])
plt.show() #exibe o histograma da imagem equalizada
```

## 5) Verifique os métodos de vizinho mais próximo (nearest), bilinear (bilinear) e bicúbica (bicubic) para realizar o redimensionamento da imagem Fig3.13.jpg. 
Para isso, reduza a imagem em 4 vezes e depois volte ao tamanho original
usando o mesmo método. Compare os resultados.

```python
from skimage.transform import rescale

im = skimage.io.imread('/content/drive/MyDrive/PDI_Labs/Lab1-Amanda/Lab-1/Fig3.13.jpg')
image_rescaled = rescale(im, 0.25, anti_aliasing=True, order = 0)
image_rescaled = rescale(image_rescaled, 4, anti_aliasing=True, order = 0)
plt.figure(figsize = (5,5))
plt.imshow(image_rescaled,cmap='gray',interpolation='nearest')
plt.show()

image_rescaled = rescale(im, 0.25, anti_aliasing=True, order = 5)
image_rescaled = rescale(image_rescaled, 4, anti_aliasing=True, order = 5)
plt.figure(figsize = (5,5))
plt.imshow(image_rescaled,cmap='gray',interpolation='nearest')
plt.show()

im = skimage.io.imread('/content/drive/MyDrive/PDI_Labs/Lab1-Amanda/Lab-1/Fig3.13.jpg')
image_rescaled = rescale(im, 0.1, anti_aliasing=True, order = 0)
image_rescaled = rescale(image_rescaled, 10, anti_aliasing=True, order = 0)
plt.figure(figsize = (5,5))
plt.imshow(image_rescaled,cmap='gray',interpolation='nearest')
plt.show()

image_rescaled = rescale(im, 0.1, anti_aliasing=True, order = 5)
image_rescaled = rescale(image_rescaled, 1, anti_aliasing=True, order = 5)
plt.figure(figsize = (5,5))
plt.imshow(image_rescaled,cmap='gray',interpolation='nearest')
plt.show()

#Possíveis valores de order:
#0: Nearest-neighbor
#1: Bi-linear (default)
#2: Bi-quadratic
#3: Bi-cubic
#4: Bi-quartic
#5: Bi-quintic
```

## 6) Use filtros sobre a imagem Fig3.40(a).jpg e verifique o resultado final.
a) Filtro de média 11x11;
b) Filtro de Laplace;

```python
from skimage import filters, morphology

im = skimage.io.imread('/content/drive/MyDrive/PDI_Labs/Lab1-Amanda/Lab-1/Fig3.40(a).jpg')
selem = morphology.rectangle(3,3)
im2 = filters.rank.mean(im, selem=selem)
plt.figure(figsize = (5,5))
plt.imshow(im2,cmap='gray')
plt.show()
```

```python
selem = morphology.rectangle(11,45)
im2 = filters.rank.mean(im, selem=selem)
plt.figure(figsize = (5,5))
plt.imshow(im2,cmap='gray')
plt.show()
```

```python
# Laplace
im2 = filters.laplace(im)
plt.figure(figsize = (5,5))
plt.imshow(im2,cmap='gray')
plt.show()
```

```python
# Gaussiano
im2 = filters.gaussian(im, sigma=10)
plt.figure(figsize = (5,5))
plt.imshow(im2,cmap='gray')
plt.show()
```

## 7) Adicione ruído sal e pimenta na imagem Fig3.13.jpg e depois use um filtro de mediana para remover. 
Teste diferentes quantidades de ruído e tamanhos de filtros.

```python
im = skimage.io.imread('/content/drive/MyDrive/PDI_Labs/Lab1-Amanda/Lab-1/Fig3.13.jpg')
plt.figure(figsize = (5,5))
plt.imshow(im,cmap='gray')
plt.show()
```

```python
rd = np.random.uniform(0,1,im.shape)
im2 = im
im2[rd >= 0.95] = 255
im2[rd <= 0.05] = 0
plt.figure(figsize = (5,5))
plt.imshow(im2,cmap='gray')
plt.show()
```

```python
selem = morphology.rectangle(5,5) # tamanho da mascara do filtro
im2 = filters.median(im, selem=selem)
plt.figure(figsize = (5,5))
plt.imshow(im2,cmap='gray')
plt.show()
```

```python
selem = morphology.rectangle(45,45) # tamanho da mascara do filtro
im2 = filters.median(im, selem=selem)
plt.figure(figsize = (5,5))
plt.imshow(im2,cmap='gray')
plt.show()
```
